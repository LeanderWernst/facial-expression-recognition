```python
from PIL import Image
import matplotlib.pyplot as plt

image1 = "/home/lndr/Schreibtisch/affnet/neutral.jpg"
image2 = "/home/lndr/Schreibtisch/affnet/happiness.jpg"
image3 = "/home/lndr/Schreibtisch/affnet/sadness.jpg"
image4 = "/home/lndr/Schreibtisch/affnet/surprise.jpg"
image5 = "/home/lndr/Schreibtisch/affnet/fear.jpg"
image6 = "/home/lndr/Schreibtisch/affnet/disgust.jpg"
image7 = "/home/lndr/Schreibtisch/affnet/anger.jpg"
image8 = "/home/lndr/Schreibtisch/affnet/contempt.jpg"

classes = {0: 'neutral', 
           1: 'happiness', 
           2: 'sadness', 
           3: 'surprise', 
           4: 'fear', 
           5: 'disgust', 
           6: 'anger', 
           7: 'contempt'}

# Define your 8 images and headlines (replace these with your actual data)
images = [image1, image2, image3, image4, image5, image6, image7, image8]
headlines = ["0: neutral", "1: happiness", "2: sadness", "3: surprise", "4: fear", "5: disgust", "6: anger", "7: contempt"]

# Load RGB images
images = [Image.open(image_path) for image_path in images]

# Create a 2x4 grid of subplots
fig, axes = plt.subplots(2, 4, figsize=(12, 6))

# Loop through the images and headlines, and plot them in the subplots
for i, (image, headline) in enumerate(zip(images, headlines)):
    row = i // 4
    col = i % 4
    ax = axes[row, col]
    ax.imshow(image)
    ax.set_title(headline)
    ax.axis('on')  # Show the axes

# Adjust spacing between subplots
plt.tight_layout()

# Save the plot as a PDF file
plt.savefig("output.pdf")

# Show the plot
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

image1 = "/home/lndr/Schreibtisch/affnet/normal.jpg"
image2 = "/home/lndr/Schreibtisch/affnet/occluded.jpg"

# Define your 8 images and headlines (replace these with your actual data)
images = [image1, image2]
headlines = ["original", "occluded"]

# Load RGB images
images = [Image.open(image_path) for image_path in images]

# Create a 2x4 grid of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Loop through the images and headlines, and plot them in the subplots
for i, (image, headline) in enumerate(zip(images, headlines)):
    row = i // 1
    col = i % 2
    ax = axes[row, col]
    ax.imshow(image)
    ax.set_title(headline)
    ax.axis('on')  # Show the axes

# Adjust spacing between subplots
plt.tight_layout()

# Save the plot as a PDF file
plt.savefig("output.pdf")

# Show the plot
plt.show()
```
